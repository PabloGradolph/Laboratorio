Los números que no están tachados (en negrita) en la tabla son precisamente los primos menores que $n$.


> Implementa una función $\verb=eratostenes(n)=$ que devuelva la lista de todos los primos menores que $n$ (esta función debería hacer lo mismo que la función de sage $\verb=prime_range(n)=$).

En esta práctica pueden ser útiles los siguientes métodos asociados a una lista L:

- $\verb=L.pop(i)=$: borra el elemento que está en posición $i$-ésima en la lista L.
- $\verb=L.remove(i)=$: borra el primer elemento igual a $i$ en la lista L.

Por ejemplo, si $\verb=L= = [3,5,7,9]$, después de la instrucción $\verb=L.pop(2)=$, tendríamos que $\verb=L= = [3,5,9]$ y después de la instrucción $\verb=L.remove(5)=$, tendríamos que $\verb=L= =[3,7,9]$.

```python
def eratostenes(n):
    primos = list(range(2,n))
    p=2
    i = 0
    while p*p<n:
        p = primos[i]
        for m in range(p^2,n,p):
            if m in primos:
                primos.remove(m)
        i = i + 1
    return primos

eratostenes(42)
```

Euclides fue un matemático griego del siglo IV a. C. que escribió el célebre tratado *Los elementos*, que recoge una gran parte de las matemáticas que conocían los griegos. Uno de los célebres teoremas que se recogen en su obra es el Teorema Fundamental de la Aritmética, que dice que todo número natural se puede factorizar de manera única como producto de números primos.

> Implementa la función $\verb=factores_primos(n)=$ que reciba un número natural $n$ y devuelva dos listas: una con los primos que divididen a $n$ (ordenados de menor a mayor) y otra con sus multiplicidades. Por ejemplo:

$\verb=factores_primos(84)= = [2,3,7],[1,2,1]$

$\verb=factores_primos(250)= = [2,5],[1,3]$

(esta función es comparable a la función de sage $\verb=factor(n)=$)


```python
def factores_primos(n):
    factores = []
    multiplicidades = []
    for p in eratostenes(n+1):
        mult = 0
        while n % p == 0:
            mult = mult + 1
            n = n//p
        if mult > 0:
            factores.append(p)
            multiplicidades.append(mult)
        if n == 1:
            break
    return factores, multiplicidades

factores_primos(54)
```

Finalmente, implementa una función que calcule el máximo común divisor de dos números usando el algoritmo estudiado en primaria (aquí no hay ninguna cita de ningún griego porque los griegos no usaban ese algoritmo).

> Implementa una función que calcule el máximo común divisor de dos números (esta función debería hacer lo mismo que la función de sage $\verb=gcd(a,b)=$).


```python
def mcd(a,b):
    Pa, Ma = factores_primos(a)
    Pb, Mb = factores_primos(b)
    mcd = 1
    while len(Pa) > 0 and len(Pb)>0:
        if Pb[0] == Pa[0]:
            mcd = mcd*Pa[0]^(min(Ma[0],Mb[0]))
            Pa.pop(0)
            Pb.pop(0)
            Ma.pop(0)
            Mb.pop(0)
        elif Pa[0] < Pb[0]:
            Pa.pop(0)
            Ma.pop(0)
        else:
            Pb.pop(0)
            Mb.pop(0)
    
    return mcd
```
