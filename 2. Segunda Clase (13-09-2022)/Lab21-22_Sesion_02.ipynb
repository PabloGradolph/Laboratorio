### **Ejercicio 1** ### 

Se considera la funcion $\displaystyle f(x)=\frac{x^2+1}{|x-4|}$.

**a)** Determinar sus asíntotas oblicuas con la ayuda de SageMath. 

**b)** Dibujar la función  junto con sus asíntotas (vertical y oblicuas) en el intervalo $[-40,40]$.

*Sugerencia:* Para dibujar la asíntota vertical puede ser útil la instrucción $\texttt{line}$.

**c)** Calcula los puntos del dominio de la función donde se anula su derivada.

```python
#Definimos la función
f(x)=(x**2+1)/abs(x-4)

#Representamos la función
show(plot(f(x),-40,40,ymin=-40,ymax=40,aspect_ratio=1))
```

```python
#Calculamos los límites para ver si hay asíntota horizontal
limit(f(x), x=oo)
limit(f(x), x=-oo)
```

```python
#No tiene asíntotas horizontales
#Calculamos los límites para las asíntotas oblicuas
m = limit(f(x)/x, x=oo)
n = limit(f(x)-m*x, x=oo)
y = m*x+n
print(f"Una asíntota oblicua es: y = {y}")

m = limit(f(x)/x, x=-oo)
n = limit(f(x)-m*x, x=-oo)
z = m*x+n
print(f"La otra asíntota oblicua es: y = {z}")
```

```python
#Calculamos la asíntota vertical
g(x) = x - 4
solve(g(x)==0, x)
t(x) = (x == 4)
```

```python
#Dibujamos la gráfica con las asíntotas
show(plot(f(x),-40, 40, ymin=-40,ymax=40,color='blue') + plot(y,-40,40,ymin=-40,ymax=40,color="red") + plot(z, -40,40,ymin=-40,ymax=40,color="green") + line([(4,-40),(4,40)],rgbcolor="black"))
```

### **Ejercicio 2** ###

    
El **Teorema de Rolle** es un conocido teorema de cálculo de una variable que dice que si $f$ es continua en el intervalo $[a,b]$, derivable en $(a,b)$ y verifica que $f(a)=f(b)$, entonces existe un punto $c\in (a,b)$ tal que $f'(c)=0$. 

El **Teorema integral del valor medio**, otro enunciado famoso de cálculo de una variable, dice que si $f$ es una función continua en el intervalo cerrado $[a,b]$, entonces existe un valor $d$ en dicho intervalo tal que $\int_a^b f(x)\,{\rm d}x=f(d)(b-a)$.  

Sea $\displaystyle f(x)=\frac{e^x\operatorname{sen} x}2$, $a=0$, $b=\pi$.

    
**a)** Dibuja  (por separado) las gráficas de la función $f$ y de su derivada en en el intervalo $[a,b]$  y calcula un valor correcto (aproximado) para el valor $c$ del Teorema de Rolle aplicado a dicha $f$ en ese intervalo. ¿Cuántos posibles valores de $c$ crees que existen? Dibuja de nuevo la gráfica de $f$ en el intervalo, añadiéndole ahora un pequeño círculo de otro color (usa para ello la instrucción *points*) que muestre el punto de la gráfica donde la derivada se anula.

*Sugerencia:* Recuerda la existencia de la función $\texttt{find}$_$\texttt{root}$.

```python
#Definimos la función
f(x)=(e**x*sin(x)/2)
f1 = diff(f(x), x)
a = 0
b = pi
show(plot(f(x), a, b, ymin=0, ymax=4, color='blue', aspect_ratio=1))
show(plot(f1, a, b, ymin=0, ymax=3, color='red', aspect_ratio=1))

#Se cumple el Teorema de Rolle para el intervalo [a,b], por tanto debe existir un c tal que f'(c)=0.
c = find_root(f1==0, a, b)
print("-------------------------------------------------------------------")
print(f"El punto c para el cual se anula la derivada es: {c}")
print("-------------------------------------------------------------------")

#Dibujamos de nuevo la gráfica de f en el intervalo, añadiéndole ahora un pequeño círculo de otro color que muestre el punto dónde se anula la derivada.
s = 2.356194490192236
y = f(2.356194490192236)
punto = (s, y)
print("Mostrando el punto por pantalla...")
show(plot(f(x), a, b, ymin=0, ymax=4, color='blue', aspect_ratio=1) + points(punto, color="purple", size=20))
```

**b)** Calcula **todos** los valores correctos (aproximados) para el $d$ del enunciado del Teorema Integral del Valor Medio aplicado a dicha $f$ en ese intervalo $[a,b]$. ¿Cuántos posibles valores de $d$ crees que existen?


*Sugerencia:* Como en el apartado a), te puede ayudar dibujar las gráficas de $f$ y de alguna función auxiliar.

```python
#Calculamos la integral de dicho intervalo
integral = integrate(f(x), x, a, b)

#Despejamos en función del Teorema Integral del valor Medio
sol = integral/(b - a)
show(plot(f(x),x, a, b) + plot(sol,x, a, b, color="red"))
#Sacamos el/los valores de d
d1 = find_root(f(x) == sol,0,2)
d2 = find_root(f(x) == sol,2.5,pi)
show('Los valores correctos son:')
show(f"{d1} y {d2}")
```

Se considera la matriz 

$$
A = \begin{pmatrix}
a&0&1\\
0&b&0\\
1&0&a\\
\end{pmatrix}.
$$

**a)** Determine los valores de los parámetros reales $a$ y $b$ para los que $A = A^{-1}$.

**b)** Para $a = b = 2$, calcule la matriz inversa de $A$.

*Sugerencia:*  En SageMath, la instrucción $\texttt{A[i,j]}$ devuelve el elemento que ocupa la posición $(i,j)$ en la matriz $A$. Es muy importante tener en cuenta que SageMath empieza a numerar las filas y columnas por el índice 0.

```python
#Sugerencia: Lista de los elementos de A perteneciente a M(mxn)
#Escribir [A[ij] for i in srange(m) for j in srange(n)]
#srange(4)=[0,1,2,3]
#[A[ij]==A⁻1[ij] for i in...]

#Definimos las variables
var('a')
var('b')

#Creamos la matriz
elementos = [[a, 0, 1], [0, b, 0], [1, 0, a]]
A = Matrix(3, 3, elementos)
B = A.inverse()
show("Matrices:")
show(A)
show(B)

#fabrico la lista de todas las igualdades:
ecs = [A[i,j] == B[i,j] for i in range(3) for j in range(3)]
show('Las soluciones son:')
show(solve(ecs,a,b))

```

Una tienda online de productos gourmet elabora tres tipos de cafés exclusivos, el Gold Cuvée (a 7.85 euros/kg),
el Paradiso (a 13.3 euros/kg) y el Cremissimo (a 24.85 euros/kg). Para ello utiliza solo dos tipos de grano, el
Arábica y el Robusta. El Gold Cuvée tiene un 90% de grano tipo Arábica, el Paradiso un 85% y el Cremissimo
un 80%.
A lo largo de un mes han necesitado utilizar 27.1 kg de grano del tipo Robusta para atender todos los pedidos y
han ingresado un total de 3112.5 euros. Sabiendo que se ha vendido doble cantidad de café Cremissimo que de
las otras dos especialidades juntas, se pide calcular los kilogramos de grano del tipo Arábica que se han utilizado
a lo largo de ese mes.

```python
# x = Gold Cuvée (7.85 euros/kg)
# y = Paradiso (13.3 euros/kg)
# z = Cremissimo (24.85 euros/kg)
# Tipos de grano: Arábica y Robusta
# - - - - - - - - - - - - - - - - - - - - - 
# DATOS:
# Cuvée --> 90% grano Arábica
# Paradiso --> 85% grano Arábica
# Cremissimo --> 80% grano Arábica
# En un mes se han utilizado 27.1 kg de grano Robusta
# Se han ingresado 3112.5 euros.
# Cremissimo = 2*(Cuvée + Paradiso)
# - - - - - - - - - - - - - - - - - - - - - 
# SE PIDE:
# ¿Kilogramos de grano del tipo Arábica en ese mes?
# - - - - - - - - - - - - - - - - - - - - - 
# SISTEMA A RESOLVER:
# 7.85x + 13.3y + 24.85z = 3112.5
# 0.10x + 0.15y + 0.2z = 27.1
# - 2x - 2y + z = 0

# Sacamos los kilogramos de cada tipo de café en ese mes.
var('x y z')
solve([7.85*x+13.3*y+24.85*z == 3112.5, 0.10*x+0.15*y+0.2*z == 27.1, -2*x-2*y+z == 0],x,y,z)
```
